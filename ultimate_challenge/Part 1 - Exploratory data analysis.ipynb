```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read logins.json and save it to login dataframe
login = pd.read_json('logins.json')
```

```python
# Print the first three rows of data
login.head(5)
```

##### Quality issue: The login time in row 3 is before the login time of row 2. Therefore, the dataset is not ordered. 

```python
# Sort the login times and reset index
login.sort_values(by='login_time',inplace=True)
login.reset_index(drop=True,inplace=True)

# Print the first and last three rows of data
print('The first three rows of data:\n',login.head(3))
print('\nThe last three rows of data:\n',login.tail(3))
```

##### It would be better if the data contained logins timestamps before 8 pm on January 1st and after 7 pm on April 13. In order to have a better understanding of login timestamp distribution, it would also be helpful if the data contained login timestamps of remaining days of April and even the rest of months in 1970.

```python
# Set login time as index and add a new column named count.
login=login.set_index('login_time')
login['count']=1

# Count number of logins based on 15-minute time intervals
login_reduced = login.resample('15T').count()
```

```python
# Plot number of counts at every 15-minute time interval and investigate the diagram
login_reduced['count'].plot(figsize=(20,8),grid=True)
plt.ylabel('Number of logins')
plt.title('Number of logins at every 15-minute time interval from 1970-01-01 to 1970-04-13')
plt.show()
```

#### There is a potential data quality issue: a 15-minute time interval in March has more than 70 logins. There are also couple of points in April with 60 or more counts. All these points are outliers and must be investigated to find if such high number of logins really happened or they are because of data errors.

```python
# Add a new column which contains only the time and save it as string. 
login_reduced['time']=login_reduced.index.time.astype('str')

# Compute the average number of logins on 15-minute time intervals across all days and plot the results
login_reduced.groupby('time')['count'].mean().plot(figsize=(20,8),grid=True)
ticks = np.arange(0,97,4)
labels=['{}:00'.format(a) for a in np.arange(0,25)]
plt.xticks(ticks,labels)
plt.xlabel('Time')
plt.ylabel('Average number of logins')
plt.title('Average number of logins on 15-minute time intervals from 1970-01-01 to 1970-04-13')
plt.show()
```

#### The plot above is showing the average number of logins starts going up after 5 pm. After going through some up and downs around midnight, it goes down until 8 am (it experiences a significant drop around 5 am). Then, again after 8 am, the login counts  increases until 11:30 am (it experiences a big jump around 10 am). After 11:30 am, the login counts eventually decreases until 5 pm.

```python
# Add a new column which contains only the weekday. 
login_reduced['weekday']=login_reduced.index.weekday

# Compute the average number of logins on each weekday and plot the results
login_reduced.groupby('weekday')['count'].mean().plot(figsize=(20,8),grid=True)
ticks = np.arange(0,7)
labels=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
plt.xticks(ticks,labels)
plt.xlabel('Weekday')
plt.ylabel('Average number of logins')
plt.title('Average number of logins on each weekday from 1970-01-01 to 1970-04-13')
plt.ylim(0,14)
plt.show()
```

#### As the week goes by, the average number of logins goes up until the peak on Saturday. Average number of logins on Sunday is less than Saturday but still higher than the rest of weekdays.

#### Another helpful diagram would be the average number of logins per month day.
